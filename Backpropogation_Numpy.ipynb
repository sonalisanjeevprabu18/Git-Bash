```python
import numpy as np

# Input and target
X = np.array([[1], [2], [3], [4]], dtype=float)
y = np.array([[0], [0], [1], [1]], dtype=float)

# Initialize weight and bias
w = np.random.randn(1, 1)
b = 0.0

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Derivative of sigmoid
def sigmoid_derivative(a):
    return a * (1 - a)

# Learning rate
lr = 0.1

# Training loop
for epoch in range(1000):
    # Forward propagation
    z = np.dot(X, w) + b
    y_hat = sigmoid(z)

    # Loss (MSE)
    loss = np.mean((y - y_hat) ** 2)

    # Backpropagation
    error = y_hat - y
    dz = error * sigmoid_derivative(y_hat)
    dw = np.dot(X.T, dz) / len(X)
    db = np.mean(dz)

    # Update parameters
    w -= lr * dw
    b -= lr * db

# Final output
print("Final Weight:", w)
print("Final Bias:", b)
print("Predicted Output:")
print(y_hat)
```
